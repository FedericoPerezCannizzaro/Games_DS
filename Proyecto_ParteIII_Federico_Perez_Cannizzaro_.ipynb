```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import sys
sys.setrecursionlimit(2000)
dataset = df[['Price','Median playtime two weeks']]
dataset = dataset.loc[dataset['Median playtime two weeks'] * dataset['Price'] != 0 ]
sampled_data = dataset.sample(n = 1500, random_state = 42)
X = sampled_data.iloc[:, [0,1]].values
plt.figure(figsize=(5,3))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

**Aprendizaje no supervisado: Grafico de dispersion**
en este grafico de dispersion podemos ver segun el tiempo de juego lo siguiente:
- El cluster rojo representa juegos con poca cantidad de tiempo de juego invertido,tambien se ve una mayor variedad dentro del eje x en cuanto a los valores de las medias de juego
-El azul representa juegos con un alto tiempo de juego medio
